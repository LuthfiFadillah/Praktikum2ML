```python
import pandas as pd

# Load Dataset
dataset_training = pd.read_csv('data_training_praktikum.csv')
dataset_training.describe(include='all')

dataset_testing = pd.read_csv('data_testing_praktikum_tanpakelas.csv')
```

```python
# Drop Columns with mean = 0
list_column = []
import numpy as np
for i in range (60):
    if (dataset_training[dataset_training.columns[i]].dtype != np.object and dataset_training[dataset_training.columns[i]].mean() == 0):
        list_column.append(dataset_training.columns[i])

dropped_dataset = dataset_training
for column in list_column:
    dropped_dataset = dropped_dataset.drop(column,axis=1)
    
```

```python
# Encoding
data_train = dropped_dataset

# Label Encoding
categorical = {"loc": {"A": 0, "B": 1, "C": 2, "D": 3, "E": 4, "F": 5, "G": 6, "H": 7, "I": 8, "J": 9},
               "struct2": {"Initial": 0, "Medial": 1, "Final": 2},
               "cit1": {"None": 0, "Citation": 1, "Self Citation": 2},
               "cit2": {"None": 0, "Beginning": 1, "Middle": 2, "End": 3},
              }
data_train.replace(categorical, inplace=True)

# Class Encoding
data_train["labelclass"] = data_train["labelclass"].astype('category')
data_train["labelclass"] = data_train["labelclass"].cat.codes

# One-Hot Encoding
data_train = pd.get_dummies(data_train, columns=["struct3", "struct1"])
data_train = data_train.drop("struct3_Example",axis=1)
print(data_train.dtypes)
data_train.head()

#dropped_dataset.head()
```
